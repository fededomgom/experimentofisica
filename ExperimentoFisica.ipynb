```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches

df = pd.read_csv('dato.csv')

#Tiempos para recorrer 100 centimetros
t100 = df['t-100']
g100 = df['g-100']
pp100 = df['pp-100']
h100 = df['h-100']

#Tiempos para caer verticalmente
tv = df['t-v']
gv = df['g-v']
ppv = df['pp-v']
hv = df['h-v']
```

```python
t100 = t100.to_numpy()
g100 = g100.to_numpy()
pp100 = pp100.to_numpy()
h100 = h100.to_numpy()

tv = tv.to_numpy()
gv = gv.to_numpy()
ppv = ppv.to_numpy()
hv = hv.to_numpy()
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
ax = fig.add_subplot(111)
intervalo = 0.04
ax.hist(t100, bins=np.arange(0.1, 1.5, intervalo), alpha = 0.4, color= 'b')
ax.hist(g100, bins=np.arange(0.1, 1.5, intervalo), alpha = 0.4, lw=1, color= 'r')
ax.hist(pp100, bins=np.arange(0.1, 1.5, intervalo), alpha = 0.4, lw=3, color= 'c')
ax.hist(h100, bins=np.arange(0.1, 1.5, intervalo), alpha = 0.4, lw=3, color= 'y')
ax.set_xlim(0, 1.5)
ax.set_ylim(0, 13)
plt.xlabel("Tiempo [s].")
plt.ylabel("Cuentas en el intervalo")
plt.suptitle('Tiempos para recorrer 100 cent√≠metros', fontsize=16)
item_azul = mpatches.Patch(color='b', label='Pelota de Tenis')
item_rojo = mpatches.Patch(color='r', label='Pelota de Golf')
item_cyan = mpatches.Patch(color='c', label='Pelota de Ping-Pong')
item_amarillo = mpatches.Patch(color='y', label='Pelota de Hockey')
plt.legend(handles=[item_azul, item_rojo, item_cyan, item_amarillo])
plt.savefig('100cm.png', bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
ax = fig.add_subplot(111)
limite_x = 1.0
ax.hist(tv, bins=np.arange(0.1, limite_x, intervalo), ls='dashed', alpha = 0.5, lw=3, color= 'b')
ax.hist(gv, bins=np.arange(0.1, limite_x, intervalo), ls='dotted', alpha = 0.5, lw=3, color= 'r')
ax.hist(ppv, bins=np.arange(0.1, limite_x, intervalo), alpha = 0.5, lw=3, color= 'c')
ax.hist(hv, bins=np.arange(0.1, limite_x, intervalo), alpha = 0.5, lw=3, color= 'y')
ax.set_xlim(0, limite_x)
ax.set_ylim(0, 17)
plt.xlabel("Tiempo [s]")
plt.ylabel("Cuentas en el intervalo")
plt.suptitle('Tiempos para caer verticalmente', fontsize=16)
item_azul = mpatches.Patch(color='b', label='Pelota de Tenis')
item_rojo = mpatches.Patch(color='r', label='Pelota de Golf')
item_cyan = mpatches.Patch(color='c', label='Pelota de Ping-Pong')
item_amarillo = mpatches.Patch(color='y', label='Pelota de Hockey')
plt.legend(handles=[item_azul, item_rojo, item_cyan, item_amarillo])

plt.show()
```
